```python
import pandas as pd
```

```python
def read_text(path):
    dataframe = pd.read_csv(path, sep="\n", header=None)
    return dataframe
```

```python
answers_path = "answers.text"
```

```python
answers_data = read_text(answers_path)
answers_data.columns = ["answers"]
```

```python
answers_data
```

```python
questions_path = "questions.text"
```

```python
questions_data = read_text(questions_path)
questions_data.columns = ["questions"]
```

```python
questions_data
```

```python
questions_and_answers = questions_data.join(answers_data)
```

```python
questions_and_answers
```

```python
questions_and_answers.to_csv('questions_and_answers.csv')
```
